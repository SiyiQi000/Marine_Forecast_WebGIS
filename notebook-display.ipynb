```python
import numpy as np
from scipy.ndimage import filters
img = np.zeros((300, 300))
img[np.random.randint(0, 300, 1000), np.random.randint(0, 300, 1000)] = 255
img2 = filters.gaussian_filter(img, 4, order=2)

import io
import matplotlib
from IPython import display
buf = io.BytesIO()
matplotlib.image.imsave(buf, img2, cmap="gray")
display.Image(buf.getvalue())
```

```python
shell = get_ipython()
png_formatter = shell.display_formatter.formatters[u'image/png']
def ndarray_formatter(array):
    import io
    import numpy as np
    from matplotlib.image import imsave
    buf = io.BytesIO()
    if array.ndim == 3 and array.shape[-1] == 3:
        imsave(buf, array)
    elif array.ndim == 2 and array.dtype == np.uint8:
        imsave(buf, array, cmap="gray", vmin=0, vmax=255)
    elif array.ndim == 2:
        imsave(buf, array, cmap="gray")
    else:
        return None
    return buf.getvalue()
png_formatter.deferred_printers[("numpy", "ndarray")] = ndarray_formatter
```
